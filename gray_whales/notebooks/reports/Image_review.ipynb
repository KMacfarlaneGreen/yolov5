```python
import pandas as pd 
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.patches as patches
%matplotlib inline
```

```python
cropped_image_path = "/gws/nopw/j04/ai4er/users/kmgreen/data/Publication_data/pub_dataset_final/val/images/sanig2_42.png"
cropped_label_path = "/gws/nopw/j04/ai4er/users/kmgreen/data/Publication_data/pub_dataset_final/val/labels/sanig2_42.txt"
```

```python
cropped_image = Image.open(cropped_image_path)
```

```python
fig, ax = plt.subplots(figsize = (20,26))
ax.xaxis.tick_top()
ax.imshow(cropped_image)

boxes = pd.read_csv(cropped_label_path, sep=" ", header=None)
boxes

for i in range(0,boxes.shape[0]):
    x = (boxes[1][i] - boxes[3][i]*0.5)*cropped_image.size[0]
    y = (boxes[2][i] - boxes[4][i]*0.5)*cropped_image.size[1]
    w = boxes[3][i] * cropped_image.size[0]
    h = boxes[4][i] * cropped_image.size[1]
    ax.add_patch(patches.Rectangle((x,y),w,h, fill=False, edgecolor='red', lw=2))

plt.show()
```
